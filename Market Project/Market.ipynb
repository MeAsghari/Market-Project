## Convert Float to Int

```python
df3 = df3.drop(['Index' , 'Date'] , axis=1)            # first of all we remove object columns from dataframe
df3
```

```python
df3 = df3.round(0)                     # step1 : round float To the nearest integer
df3 = df3.astype('int64')              # step2 : Convert float to integer
df3
```

```python
df3.columns                    # check the columns
```

```python
# Now We need to add the two deleted columns to the dataframe
df3[['Index' , 'Date']] = df2[['Index' , 'Date']]
columns_title = ['Index' , 'Date' , 'Open', 'High', 'Low', 'Close', 'Adj Close']
df3 = df3.reindex(columns = columns_title)
df3
```

## Price-Time Chart ÙŽAnd Linear ÙŽÙŽAnalysis

```python
# Price-Time Chart And Linear Analysis
plt.figure(figsize=(12 , 8))
plt.plot(df3['Date'] , df3['Adj Close'], color='b')
plt.xlabel('Date' , color = 'r' , fontsize=15)
plt.xticks(rotation=90)
plt.ylabel('Adj Close', color = 'r' , fontsize=15)
plt.title('Price-Time Chart' , color ='r' , fontsize=20)
plt.grid()
plt.show()
```
